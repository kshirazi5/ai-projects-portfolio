# Time Series Revenue Trend Analysis

This project simulates 24 months of revenue data for a business and visualizes trends over time. The data shows monthly revenue growth, helping stakeholders identify seasonal patterns and overall performance.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate data
np.random.seed(0)
months = pd.date_range(start='2022-01-01', periods=24, freq='M')
revenue = np.random.randint(10000, 20000, size=24) + np.linspace(0, 5000, 24)

df = pd.DataFrame({'Month': months, 'Revenue': revenue})
```

## Monthly Revenue Over Time

We use Matplotlib to create a line chart that visualizes revenue performance from January 2022 to December 2023. This type of analysis is essential for identifying growth trends and planning budgets.


```python
# Plot revenue trend
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Revenue'], marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
